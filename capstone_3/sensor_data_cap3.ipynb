So we have 3 status, and we need to plot the statuses next to the meters to give us a suggestion of what/when the sensors will fail or break.  For simplicity sake, we can assume that 25% of the data could be used to train the model and measure 2 BROKEN states, the remaining 75% of the data is used to test the predictability of the model based on input parameters (covers 5 BROKEN points).

Let's get a graphical representation of what each sensor is doing right before the broken status. We'll assume the BROKEN state is a value of 0, RECOVERING is 0.5, and NORMAL is 1.  We'll make a new column and plot this as ops.

```python
ops = [(df['machine_status'] =='BROKEN'), (df['machine_status'] =='NORMAL'), (df['machine_status'] =='RECOVERING')]
choices = [0, 1, 0.5]
df['Operation'] = np.select(ops, choices, default=0)
```
