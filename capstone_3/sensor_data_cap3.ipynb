So we have 3 status, and we need to plot the statuses next to the meters to give us a suggestion of what/when the sensors will fail or break.  For simplicity sake, we can assume that 25% of the data could be used to train the model and measure 2 BROKEN states, the remaining 75% of the data is used to test the predictability of the model based on input parameters (covers 5 BROKEN points).

Let's get a graphical representation of what each sensor is doing right before the broken status. We'll assume the BROKEN state is a value of 0, RECOVERING is 0.5, and NORMAL is 1.  We'll make a new column and plot this as ops.

```python
ops = [(df['machine_status'] =='BROKEN'), (df['machine_status'] =='NORMAL'), (df['machine_status'] =='RECOVERING')]
choices = [0, 1, 0.5]
df['Operational'] = np.select(ops, choices, default=0)
```

```python
df.plot(subplots=True, sharex=True, figsize = (20,50))
```

Just by looking at these plots compared to the broken/normal/recovering we can see that sensors 38-49 all seem to have big spikes around the times the operation drops to broken.  Sensors 1,2,3 seem to have a lower rate/level when the recovery stage is happening.

```python
# features extractor for columns
def get_column_features(df):
    
    df_features = pd.DataFrame()
    col_list = ['Col','Max','Min','q_0.25','q_0.50','q_0.75']
    
    for column in df:
        tmp_q = df[column].quantile([0.25,0.5,0.75])
        tmp1 = pd.Series([column,max(df[column]),min(df[column]),tmp_q[0.25],tmp_q[0.5],tmp_q[0.75]],name='d')
        tmp1.index = col_list
        df_features = df_features.append(tmp1)
        
    return df_features

# get list of signal columns without timestamp and status:
signal_columns = [c for c in df.columns if c not in ['timestamp', 'machine_status']]
get_column_features(df[signal_columns])
```

```python
#locate indices of failure events and recovering and normal state
normal_index = df.loc[df['machine_status'] == 'NORMAL'].index
failure_index = df.loc[df['machine_status'] == 'BROKEN'].index
recovering_index = df.loc[df['machine_status'] == 'RECOVERING'].index

bef_failure_index = list()
for j in failure_index:
    for i in range(24*60):
        bef_failure_index.append(j-i)

bef_failure_index.sort()

#locate timestamps of failures:
failures_timestamps = df.loc[failure_index,'timestamp']
print(failures_timestamps)
```
